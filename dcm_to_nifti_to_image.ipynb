2. 변환 과정에서 생기는 nii.gz 파일 이미지화

```python
import nibabel as nib
import matplotlib.pyplot as plt

nii = nib.load(r'C:\Users\Admin\Downloads\lung\manifest_convert\4.nii.gz').get_fdata()
print('shape of nii :', nii.shape)

for i in range(20):
    plt.subplot(5,4, i+1)
    plt.imshow(nii[:, :, i], cmap='gray')
    plt.gcf().set_size_inches(40, 40)
plt.show()
```
